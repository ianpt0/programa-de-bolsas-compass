#### Em Python, declare e inicialize uma lista contendo 250 inteiros obtidos de forma aleatória. Após, aplicar o método reverse sobre o conteúdo da lista e imprimir o resultado.

```python
import random

# Cria uma lista com 250 inteiros aleatórios entre 0 e 1000
lista = [random.randint(0, 1000) for _ in range(250)]

# Inverte a lista
lista.reverse()

# Imprime a lista
for numero in lista:
    print(numero)
```

# Section 4 | Exercícios Batch Spark
#### Em Python, declare e inicialize uma lista contendo o nome de 20 animais. Ordene-os em ordem crescente e itere sobre os itens, imprimindo um a um (você pode utilizar list comprehension aqui).  Na sequência, armazene o conteúdo da lista em um arquivo de texto, um item em cada linha, no formato CSV.

```python
# Declaração e inicialização da lista
animais = ['Cão', 'Gato', 'Elefante', 'Zebra', 'Leão', 'Tigre', 'Rato', 'Baleia', 'Golfinho', 'Tubarão', 'Águia', 'Galinha', 'Cavalo', 'Vaca', 'Porco', 'Pinguim', 'Urso', 'Girafa', 'Macaco', 'Crocodilo']

# Ordenação da lista
animais.sort()

# Iteração e impressão dos itens
for animal in animais:
    print(animal)

# Escrita do conteúdo da lista em um arquivo de texto no formato CSV
with open('animais.csv', 'w') as f:
    for animal in animais:
        f.write(f'{animal}\n')
```
