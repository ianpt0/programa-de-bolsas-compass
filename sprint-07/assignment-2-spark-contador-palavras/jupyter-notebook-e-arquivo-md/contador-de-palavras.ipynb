```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import split, explode, col

# Inicialize um SparkSession
spark = SparkSession.builder \
    .appName("Contador de Palavras") \
    .getOrCreate()

# Carregue o arquivo de texto
caminho_arquivo = "README.md"
df = spark.read.text(caminho_arquivo)

# Divida cada linha em palavras e conte a ocorrência de cada palavra
contagem_palavras = df.select(explode(split(df.value, " ")).alias("palavra")) \
    .groupBy("palavra") \
    .count()

# Exiba as contagens
contagem_palavras.show(contagem_palavras.count(), truncate=False)


# Pare o SparkSession
spark.stop()
```
